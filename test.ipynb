```python
# Tells Jupyter to automatically reload imported files every time you run a cell, so updates are always live.
%load_ext autoreload
%autoreload 2

from tiles import Tile, Suit, HONOR_TILES, WIND_TILES, DRAGON_TILES
```

```python
# Build a shuffled pool of tiles and draw a random hand of 13 tiles
from random import shuffle
pool = []

# Add 4 copies of each numbered tile in each suit
for suit in (Suit.Tiao, Suit.Wan, Suit.Tong):
    for value in range(1, 10):
        pool.extend(Tile(value, suit) for _ in range(4))  # Add 4 distinct Tile objects

# Add 4 copies of each honor tile
for value in HONOR_TILES:
    if value in WIND_TILES:
        pool.extend(Tile(value, Suit.Fang) for _ in range(4))  # Add 4 distinct Tile objects
    if value in DRAGON_TILES:
        pool.extend(Tile(value, Suit.Dragon) for _ in range(4))  # Add 4 distinct Tile objects

shuffle(pool)
hand = pool[:13]

print("--- RANDOM HAND (UNSORTED) ---")
for tile in hand:
    print(tile)

hand.sort()

print("\n--- SORTED HAND ---")
for tile in hand:
    print(tile)
```
